```python
def asciiCheck(ser):
    listy = []
    for i in ser:
        listy.append(all(ord(c) < 128 for c in i))
    ser = pd.Series(listy, copy=False)
    return ser
```

```python
def asciiCheckWrong(ser):
    listy = []
    for i in ser:
        listy.append(all(ord(c) < 300 for c in i))
    ser = pd.Series(listy, copy=False)
    return ser
```

```python
videoDF = videoDF[asciiCheck(videoDF["_id"])].reset_index(drop=True)
videoDF = videoDF[asciiCheck(videoDF["type"])].reset_index(drop=True)
videoDF = videoDF[asciiCheck(videoDF["title"])].reset_index(drop=True)
videoDF = videoDF[asciiCheck(videoDF["director"])].reset_index(drop=True)
videoDF = videoDF[asciiCheck(videoDF["cast"])].reset_index(drop=True)
videoDF = videoDF[asciiCheck(videoDF["country"])].reset_index(drop=True)
videoDF = videoDF[asciiCheck(videoDF["date_added"])].reset_index(drop=True)
videoDF = videoDF[asciiCheck(videoDF["rating"])].reset_index(drop=True)
videoDF = videoDF[asciiCheck(videoDF["duration"])].reset_index(drop=True)
videoDF = videoDF[asciiCheck(videoDF["listed_in"])].reset_index(drop=True)
videoDF = videoDF[asciiCheck(videoDF["description"])].reset_index(drop=True)
len(videoDF)
```

```python
videoDF
```

```python
videoDF['date_added'] = pd.to_datetime(videoDF['date_added'], errors='coerce')
videoDF.head()
```

```python
videoDF[videoDF["date_added"].isna()]
```

```python
jsonbook = videoDF.to_json(orient='records', date_format="iso")
parsed = json.loads(jsonbook)
jsonfinal = json.dumps(parsed, indent=4) 
with open("videos.json",'w') as outfile:
    outfile.write(jsonfinal)
```

```python
audioDF = pd.read_csv("Audible_Dataset_final.csv")
audioDF = audioDF[["_id","Book Title", "Book Subtitle", "Book Author", "Book Narrator", "Audio Runtime", "Audiobook_Type", "Categories", "Rating"]]
audioDF['available'] = 1
audioDF
```

```python
audioDF = audioDF.dropna().reset_index(drop=True)
len(audioDF)
```

```python
audioDF = audioDF[asciiCheck(audioDF["Book Title"])].reset_index(drop=True)
audioDF = audioDF[asciiCheck(audioDF["Book Subtitle"])].reset_index(drop=True)
audioDF = audioDF[asciiCheck(audioDF["Book Author"])].reset_index(drop=True)
audioDF = audioDF[asciiCheck(audioDF["Book Narrator"])].reset_index(drop=True)
audioDF = audioDF[asciiCheck(audioDF["Audio Runtime"])].reset_index(drop=True)
audioDF = audioDF[asciiCheck(audioDF["Audiobook_Type"])].reset_index(drop=True)
audioDF = audioDF[asciiCheck(audioDF["Categories"])].reset_index(drop=True)
len(audioDF)
```

```python
audioDF
```

```python
jsonbook = audioDF.to_json(orient='records', date_format="iso")
parsed = json.loads(jsonbook)
jsonfinal = json.dumps(parsed, indent=4) 
with open("audiobooks.json",'w') as outfile:
    outfile.write(jsonfinal)
```
