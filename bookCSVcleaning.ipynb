```python
bookdb = pd.read_csv(f"{path}/bookcleaningpart1.csv")
```

```python
def asciiCheck(ser):
    listy = []
    for i in ser:
        listy.append(all(ord(c) < 128 for c in i))
    ser = pd.Series(listy, copy=False)
    return ser
```

```python
bookdb = bookdb.reset_index(drop=True)
bookdb
```

```python
bookdb = bookdb[asciiCheck(bookdb["title"])].reset_index(drop=True)
bookdb = bookdb[asciiCheck(bookdb["author"])].reset_index(drop=True)
bookdb = bookdb[asciiCheck(bookdb["language_code"])].reset_index(drop=True)
bookdb = bookdb[asciiCheck(bookdb["publication_date"])].reset_index(drop=True)
bookdb = bookdb[asciiCheck(bookdb["publisher"])].reset_index(drop=True)
```

```python
bookdb
```

```python
bookdb.to_csv(f"{path}/bookcleaningpart2.csv", index=False)
```
