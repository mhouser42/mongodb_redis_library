```python
import json
import random
import re
import secrets
from faker import Faker
import pandas as pd
import pymongo as pym

#store the ID as underscore ID
```

```python
client = pym.MongoClient("mongodb+srv://<username>:<password>@librarydb.tec4i8b.mongodb.net/?retryWrites=true&w=majority")
db = client['libraryDB']
employees = db['employees']

schedules = ["A", "B", "C", "D", "E"]
wagetype = ["Salary","Hourly"]
```

```python
def getSch(wage_type):
    if wage_type == 'Salary':
        return random.choice(['A', 'C'])
    else:
        return random.choice(['B', 'D', 'E'])
getSch('Hourly')
```

```python
def getWageT():
    return random.choice(wagetype)
getWageT()
```

```python
def getMoney(wage_type, schedule):
    if wage_type == "Salary":
        return round(random.uniform(35000,69696.96,),2)
    elif schedule in ('A', 'C', 'E'):
        return round(random.uniform(35000, 55555.95,),2)
    else:
        return round(random.uniform(20000, 40000.95,),2)
getMoney('Hourly', 'C')
```

```python
def gen_phone():
    first = str(random.randint(100,999))
    second = str(random.randint(1,888)).zfill(3)

    last = (str(random.randint(1,9998)).zfill(4))
    while last in ['1111','2222','3333','4444','5555','6666','7777','8888']:
        last = (str(random.randint(1,9998)).zfill(4))
        
    return '{}-{}-{}'.format(first,second, last)
gen_phone()
```

```python
def randomHours(this_schedule):
    if this_schedule in ('A', 'C'):
        return 40
    elif this_schedule == 'E':
        return random.randint(30, 40)
    else:
        return random.randint(10,30)
randomHours('D')
```

```python
def get_email(given, surname):
    username = f'{given[:2]}{surname}'.lower()
    urls = ['gmail', 'yahoo', 'live']
    return username, f'{username}@{random.choice(urls)}.com'
```

```python
fake = Faker()
data = []
for i in range(50):
    surname = fake.last_name()
    given = fake.first_name()
    username, email = get_email(given, surname)
    wage_type = getWageT()
    schedule = getSch(wage_type)
    this_employee = { 
            "_id": i+1
            ,"surname": surname
            ,"firstName": given
            ,"schedule": schedule
            ,"weeklyHours": randomHours(schedule)
            ,"wageType": wage_type
            ,"pay": getMoney(wage_type, schedule)
            ,"phone": gen_phone()
            ,"email": email
            ,"username": username
            ,"password": secrets.token_urlsafe(random.randint(7,20))
        }
    data.append(this_employee)
    employees.insert_one(this_employee)
empDF = pd.DataFrame(data)
```
